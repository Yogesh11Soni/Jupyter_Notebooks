```python
@ray.remote
class person():
    def __init__(self) -> None:
        self.age = random.randint(10,80)
        self.name = random.choice(['jon','susan','frank','anne','ashley','bill'])
    
    def say_hello(self):
        return f'age is {self.age} and name is {self.name}'
    
    def grow_older(self):
        self.age += 1 
        return self.age

    
```

```python
people =  [person.remote() for i in range(10)]
```

```python
ray.wait(people[0])
```

```python
ray.get(people[0].say_hello.remote())
```

```python
for person in people:
    print(ray.get(person.say_hello.remote()))
```
