```python
import time
db1 = client['bulk']
collection1 = db1['bulk_update']

bulk_data = [{'_id':i,'name':f'test{i}','time':int(time.time())} for i in range(10000)]

collection1.insert_many(bulk_data)
```

```python
for i in collection1.find({'name':'test395'}):
    print(i)
collection1.find_one({'_id':45})
```
